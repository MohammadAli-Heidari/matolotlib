```python
import matplotlib.pyplot as plt
```

```python
x = [2,5,9]
y = [6,14,17]
y2 = [4,11,15]
plt.plot(x,y,color = 'r',linewidth=3)
plt.plot(x,y2,color = 'b',linewidth=3)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Chart Test\n Nemoodar')
plt.legend()
plt.show()
```

```python
age = [25,16,24,13,19,9,38,37,44,41]
bin = [0,10,20,30,40,50]
plt.hist(age , bin , histtype = 'bar' , rwidth = 0.9)
plt.show()
```

```python
x = [25,16,24,13,19,9]
y = [0,10,20,30,40,50]
plt.scatter(x,y,label = 'test' , marker = '*' , s = 100);
```

```python
major = ['eng' , 'medic' , 'art' , 'economy']
stu = [50,100,80,55]
plt.pie(stu , labels = major)
plt.title('UNIVERSITY')
plt.show()
```

```python
x = [2,4,6,8,10,12]
y1 = [0,10,20,30,40,50]
y2 = [5,15,10,40,65,70]
plt.style.use('seaborn')
plt.plot(x,y1,label='First')
plt.plot(x,y2,label='second')
plt.legend()
plt.fill_between(x,y1,y2,alpha = 0.1)
plt.figure(figsize=(25,4))
plt.show()
```
